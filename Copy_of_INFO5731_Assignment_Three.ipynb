```
# Write your code here
#Question 1.1
import pandas as pd
import nltk
from nltk.util import ngrams
from collections import Counter
import re
Punctuations_removal=[]
df=pd.read_csv("/content/sample_data/Amazon_reviews (2).csv")
for a in df['Customer_Reviews']:
   lower=a.lower() 
   lines = re.sub(r'[^\w\s]','',lower)
   num_remo=re.sub('[0-9]', '', lines)
   Punctuations_removal.append(num_remo.replace('\n', ''))
df1=pd.DataFrame(Punctuations_removal,columns=["Customer_Reviews"])
p=[]
for data in df1['Customer_Reviews']:
  n_grams = ngrams(nltk.word_tokenize(data),3)
  for grams in n_grams:
      p.append(' '.join(grams))
counters=Counter(p)
print(counters)
```

```
#Question 1.2
bigrams=[]
keys=[]
splitted_keys=[]
words_split=[]
count_values=[]
from nltk.collocations import BigramCollocationFinder
for data in df1['Customer_Reviews']:
  n_grams = ngrams(nltk.word_tokenize(data),2)
  for grams in n_grams:
      bigrams.append(' '.join(grams))
counts=Counter(bigrams)
for a in counts.items():
  keys.append(a[0])
  splitted_keys.append(a[0].split(' '))




```

```
for words in splitted_keys:
  words_split.append(words[0])
space_removal=list(filter(None,words_split))
for a in splitted_keys:
  count=0
  for temp in space_removal:
    if a[0]==temp:
      count+=1
  count_values.append(count)
for temp in range(len(list(counts.values()))):
  print(f"{keys[temp]} : {list(counts.values())[temp]/ count_values[temp]}")
```

```
#Question 1.3
from textblob import TextBlob
all_nouns=[]
probability=[]
prob=[]
for a in df1["Customer_Reviews"]:
  blob = TextBlob(a)
  for nouns in blob.noun_phrases:
    all_nouns.append(nouns)
for temp in range(len(all_nouns)):
  probability=[]
  for sentence in range(len(df1["Customer_Reviews"])):
    phrase_frq=df1["Customer_Reviews"][sentence].count(all_nouns[temp])
    phrase_freq_df=all_nouns.count(all_nouns[temp])
    probability.append(phrase_frq/phrase_freq_df)
  prob.append(probability)
```

```
#DataFrame Creation for the results
review_list=[]
for rev in range(80):
  review_list.append(f'review : {rev}')
df_final=pd.DataFrame(review_list,columns=['Reviews'],index=None)
cou=0
for result in prob:
  df_final[all_nouns[cou]]=pd.DataFrame(result)
  cou+=1
```

```
# Write your code here
#Question 2.1
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
final_Sentences=[]
ps = PorterStemmer()
wordnet=WordNetLemmatizer()
for i in range(len(df1['Customer_Reviews'])):
    sentences = df1['Customer_Reviews'][i].split()
    reviews=[]
    for word in sentences:
      if not word in set(stopwords.words('english')):
        reviews.append(wordnet.lemmatize(word))
        review = ' '.join(reviews)
    final_Sentences.append(review)
#Model 
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
X = cv.fit_transform(final_Sentences).toarray()
df5=pd.DataFrame(review_list,columns=['Reviews'],index=None)
df6=pd.DataFrame(X,columns=cv.get_feature_names())
df_final_tfidf=pd.merge(df5,df6,left_index=True, right_index=True)
df_final_tfidf
```
