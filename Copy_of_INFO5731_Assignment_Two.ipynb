(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(6) Lemmatization.

```
# Write your code here

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
benepar.download('benepar_en2')

```

```

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer 
import re
stemmer=PorterStemmer()
lemmatizer = WordNetLemmatizer() 
stop_words=stopwords.words("english")
sentence_filter_stemming=[]
sentence_filter_Lemmatization=[]
final_sentence_stemming=[]
final_sentence_lemmatize=[]
#Removing noise, such as special characters and punctuations
df['Customer_Reviews'] = df['Customer_Reviews'].str.replace(r"\W", " ").str.strip()#To remove special characters and punctuations
df['Customer_Reviews']= df['Customer_Reviews'].str.replace(r'\d+',"")#To remove Numbers
for a in df['Customer_Reviews']:
    splitting_words=word_tokenize(a)
    for b in splitting_words:
        if b not in stop_words:#Removing Stop Words
            stemming=stemmer.stem(b.lower())#Coverting text to lower case and perfroming stemming
            sentence_filter_stemming.append(stemming)
            lemmatization_words=lemmatizer.lemmatize(b.lower())#Lemmatization
            sentence_filter_Lemmatization.append(lemmatization_words)
    final_sentence_stemming.append(' '.join(sentence_filter_stemming))
    final_sentence_lemmatize.append(' '.join(sentence_filter_Lemmatization))
    sentence_filter_stemming.clear()
    sentence_filter_Lemmatization.clear()
df['Stemming']=pd.DataFrame(final_sentence_stemming)
df['Lemmatization']=pd.DataFrame(final_sentence_lemmatize)
df



    
```

```
#Question 3.2
#Dependency Parsing
import spacy
from nltk import Tree
import en_core_web_sm
en_nlp = spacy.load("en")

def Dependency_trees(tree_node):
    if tree_node.n_lefts + tree_node.n_rights > 0:
        return Tree(tree_node.orth_, [Dependency_trees(child) for child in tree_node.children])
    else:
        return tree_node.orth_
for temp in df['Lemmatization']:
        doc = en_nlp(temp)
        try:
          for sent in doc.sents:
            print(sent)
            Dependency_trees(sent.root).pretty_print()
        except:
          continue


#[to_nltk_tree(sent.root).pretty_print() for sent in doc.sents]
  
```

```
#Question 3.3
#Named Entity Recognition:
import spacy

data=spacy.load("en_core_web_sm")
final=[]
for k in df['Lemmatization']:
    person=0
    organization=0
    date=0
    location=0
    doc=data(k)
    types=[(X.text,X.label_) for X in doc.ents]
    print(types)
    if types==[]:
      final.append("No such ")
    else:
      for x,y in types:
        if y=='PERSON':
            person+=1
        if y=='ORG':
            organization+=1
        if y=='LOCATION':
            location+=1
        if y=='DATE':
            date+=1
        dat=f"{organization},{date},{location},{person}"
        final.append(str(dat))
```
